```python
list=[-21,34,-3490,0,23,1,56,-93,-95]
def positive(list):
    evenlist=[]
    oddlist=[]
    for x in list:
        if x>=0:
            evenlist.append(x)
        else:
            oddlist.append(x)
    print(evenlist)
    print(oddlist)
positive(list)
```
